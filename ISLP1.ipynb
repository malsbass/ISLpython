```python
import pandas as pd
import matplotlib.pyplot as plt 
college = pd.read_csv("college.csv")
college.head()
```

```python
college2 = pd.read_csv('College.csv', index_col=0)
college2.head()
college = college2
```

```python
college.describe()
```

```python
columns = ["Top10perc", "Apps", "Enroll"]
pd.plotting.scatter_matrix(college[columns])
plt.show()
```

```python
college.boxplot(column = ['Outstate'], by= 'Private')
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'], [0,50,100],labels=['No', 'Yes'])
college.head()
```

```python
college.value_counts('Elite')
```

```python
college.boxplot(column='Outstate', by= 'Elite')
plt.show()
```

```python
columns = ['Personal', 'PhD']
college1 = college[columns]
college1.head()
college1.plot.hist( bins = 15)
plt.show()
```

```python
auto = pd.read_csv("Auto.csv")
auto
```

```python
auto.dropna()
auto.describe()
```

```python
autoquant = auto[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']]
for column in autoquant:
    ranger = autoquant[column].max() - autoquant[column].min()
    print (f'Range of {column} = {ranger}')
```

```python
auto.drop(auto.index[10:85], axis = 0, inplace = True)
autoquant = auto[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']]
for column in autoquant:
    ranger = autoquant[column].max() - autoquant[column].min()
    print (f'Range of {column} = {ranger}')
```
